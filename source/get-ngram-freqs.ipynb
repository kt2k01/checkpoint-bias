```python
import json
import requests
import time
import csv
from tqdm import tqdm
```

```python
# read in professions, strip the article

prof923 = []

with open('/content/drive/MyDrive/checkpoint-bias/data/923-professions.txt', 'r') as f:
    for line in f:
        # with article
        profa = line.strip('\n')

        if profa[1] == ' ':
            # a
            prof923.append(profa[2:])
        else:
            # an
            prof923.append(profa[3:])

# preview
print(prof923[-10:])
```

```python
# the year range, get as large as possible, can truncate later
start_year = 1500
end_year = 2019

with open('/content/drive/MyDrive/checkpoint-bias/data/ngram-freqs.jsonl', 'w') as f:

    # loop over 923 professions
    for prof in tqdm(prof923):

        # do not send requests too often
        time.sleep(1)

        try:
            # send request
            res = requests.get(("https://books.google.com/ngrams/json?"
                                f"content={prof.replace(' ', '+')}&"
                                f"year_start={start_year}&year_end={end_year}&" # get the full year range
                                "case_insensitive=true"))
        except:
            print(f'Request for \"{prof}\" failed!')
        else:
            # write the result to file
            j = res.json()
            f.write(str(j) + '\n')
```

```python
import json
```

```python
# read in the 60 professions
# word list from https://aclanthology.org/2022.naacl-main.122.pdf

prof60 = []

with open('/content/drive/MyDrive/checkpoint-bias/data/60-professions.txt', 'r') as f:
    for line in f:
        prof60.append(line.strip('\n'))

print(prof60)
```

```python
with open('/content/drive/MyDrive/checkpoint-bias/data/professions_prompts.json', 'r') as f:
    j = json.load(f)

prof893_raw = list(j.keys())
```

```python
prof893 = []

# strip parenthesis
for prof_raw in prof893_raw:
    if '(' in prof_raw:
        # split by parenthesis and discard an extra space
        prof = prof_raw.split('(')[0][:-1]
    else:
        # no extra processing
        prof = prof_raw

    prof893.append(prof)

print(prof893)
```

```python
# count number of non-overlapping professions

t = 0

for prof in prof60:
    if prof not in prof893:
        print(prof)
        t += 1

# the final result is 33
print(f"In prof60 but not in prof893: {t}.")
```

```python
# prepend articles to the 60 professions

# 'a' stands for article
prof60a = []

for prof in prof60:
    if prof[0] in 'aeiou':
        profa = 'an ' + prof
    else:
        profa = 'a ' + prof
    prof60a.append(profa)

print(prof60a)
```

```python
# do the same for prof893

prof893a = []

for prof in prof893:
    if prof[0] in 'aeiou':
        profa = 'an ' + prof
    else:
        profa = 'a ' + prof
    prof893a.append(profa)

print(prof893a[:20])
```

```python
# again, count number of non-overlapping professions

t = 0

for profa in prof60a:
    if profa not in prof893a:
        print(profa)
        t += 1

# The final result is 33, in agreement with the previous result.
print(f"In prof60a but not in prof893a: {t}.")
```

```python
# just save all professions in the same list
# there are 3 repeated entries in 893, so the filtered length is 890
# then the total length would be 890 + 33 = 923
# if we want to get the 60 professions for male/female comparison
# just use the df filter functions to do so

prof923a = list(set(prof893a + prof60a))

print(len(prof923a))

with open('/content/drive/MyDrive/checkpoint-bias/data/923-professions.txt', 'w') as f:
    for profa in prof923a:
        f.write(profa + '\n')
```
